```python
df["Neighborhood"] = df["Neighborhood"].fillna(df["Neighborhood"].mode()[0])
```

```python
df["LotArea"] = df["LotArea"].fillna(df["LotArea"].median())
```

```python
df["OverallQual"] = df["OverallQual"].fillna(df["OverallQual"].median())
```

```python
df["OverallCond"] = df["OverallCond"].fillna(df["OverallCond"].median())
```

```python
df["HouseStyle"] = df["HouseStyle"].fillna(df["HouseStyle"].mode()[0])
```

```python
df["GarageCars"] = df["GarageCars"].fillna(0)
```

```python
threshold = len(df) * 0.5
df = df.dropna(axis=1, thresh=threshold)

print("Updated Dataset Information:")
df.info()
```
